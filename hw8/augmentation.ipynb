## Image Augmentation 

This is a simple image augmentation example using https://github.com/codebox/image_augmentor 

This tool scans a directory containing image files, and generates new images by performing a specified set of augmentation operations on each file that it finds.

Run the utility from the command-line as follows:

    python main.py <image dir> <transform1> <transform2> ...

The `<image dir>` is the path to the directory image files to be augmented.

The `transform` arguments determine what types of transformation operations are applied.


|Code|Description|Example Values|
|---|---|------|
|`fliph`|Horizontal Flip|`fliph`|
|`flipv`|Vertical Flip|`flipv`|
|`noise`|Adds random noise to the image|`noise_0.01`,`noise_0.5`|
|`rot`|Rotates the image by the specified amount|`rot_90`,`rot_-45`|
|`trans`|Shifts the pixels of the image by the specified amounts in the x and y directions|`trans_20_10`,`trans_-10_0`|
|`zoom`|Zooms into the specified region of the image, performing stretching/shrinking as necessary|`zoom_0_0_20_20`,`zoom_-10_-20_10_10`|
|`blur`|Blurs the image by the specified amount|`blur_1.5`|

This notebook will demonstrate the transformations


```python
%matplotlib inline

from PIL import Image
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def subplotImage(image,ax):
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.axis('off')
    ax.imshow(image)
```

```python
img = Image.open('/demo/aug_images/bird.png')
img = np.array(img)
plt.axis('off')
plt.imshow(img)
plt.show()
```

### Operations

#### Horizontal Flip
Mirrors the image around a vertical line running through its center

    python main.py image_dir fliph

```python


%run /demo/image_augmentor/main.py /demo/aug_images  fliph 

img2 = Image.open('/demo/aug_images/bird__fliph.png')
img2 = np.array(img2)

# show the images
f, axarr = plt.subplots(1,2)
f.set_figheight(8)
f.set_figwidth(8)


#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])

plt.tight_layout()
plt.show()
```

#### Vertical Flip
Mirrors the image around a horizontal line running through its center

    python main.py image_dir flipv

```python
%run /demo/image_augmentor/main.py /demo/aug_images  flipv 

img2 = Image.open('/demo/aug_images/bird__flipv.png')
img2 = np.array(img2)

# show the images
f, axarr = plt.subplots(1,2)
f.set_figheight(8)
f.set_figwidth(8)


#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])

plt.tight_layout()
plt.show()
```

#### Noise
Adds random noise to the image. The amount of noise to be added is specified by a floating-point numeric value that is included in the argument.

    python main.py image_dir noise_0.01 noise_0.02 noise_0.05 

```python
%run /demo/image_augmentor/main.py /demo/aug_images  noise_0.01 noise_0.02 noise_0.05 
img2 = Image.open('/demo/aug_images/bird__noise0.01.png')
img3 = Image.open('/demo/aug_images/bird__noise0.02.png')
img4 = Image.open('/demo/aug_images/bird__noise0.05.png')
# show the images
#plt.figure(figsize=(100,100))
f, axarr = plt.subplots(1,4)
f.set_figheight(10)
f.set_figwidth(10)
#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])
subplotImage(img3,axarr[2])
subplotImage(img4,axarr[3])
plt.tight_layout()
plt.show()
```

#### Rotate
Rotates the image. The angle of rotation is specified by a integer value that is included in the argument.  This value can be positor or negative.

    python main.py image_dir rot_90 rot_180 rot_-90

```python
%run /demo/image_augmentor/main.py /demo/aug_images  rot_90 rot_180 rot_-90 
img2 = Image.open('/demo/aug_images/bird__rot-90.png')
img3 = Image.open('/demo/aug_images/bird__rot180.png')
img4 = Image.open('/demo/aug_images/bird__rot90.png')
# show the images
#plt.figure(figsize=(100,100))
f, axarr = plt.subplots(1,4)
f.set_figheight(10)
f.set_figwidth(10)
#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])
subplotImage(img3,axarr[2])
subplotImage(img4,axarr[3])
plt.tight_layout()
plt.show()
```

#### Translate
Performs a translation on the image. The size of the translation in the x and y directions are specified by integer values that are included in the argument

    python main.py image_dir trans_20_20 trans_0_100 

```python
%run /demo/image_augmentor/main.py /demo/aug_images  trans_20_20 trans_0_100 
img2 = Image.open('/demo/aug_images/bird__trans20_20.png')
img3 = Image.open('/demo/aug_images/bird__trans0_100.png')


# show the images
#plt.figure(figsize=(100,100))
f, axarr = plt.subplots(1,3)
f.set_figheight(10)
f.set_figwidth(10)
#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])
subplotImage(img3,axarr[2])
plt.tight_layout()
plt.show()
```

#### Zoom/Stretch
Zooms in (or out) to a particular area of the image. The top-left and bottom-right coordinates of the target region are
specified by integer values included in the transform argument. By specifying a target region with an aspect ratio that
differs from that of the source image, stretching transformations can be performed.


    python main.py image_dir zoom_150_0_300_150 zoom_0_50_300_150 zoom_200_0_300_300 

```python
%run /demo/image_augmentor/main.py /demo/aug_images  zoom_150_0_300_150 zoom_0_50_300_150 zoom_200_0_300_300  
img2 = Image.open('/demo/aug_images/bird__zoom150_0_300_150.png')
img3 = Image.open('/demo/aug_images/bird__zoom0_50_300_150.png')
img4 = Image.open('/demo/aug_images/bird__zoom200_0_300_300.png')
# show the images
#plt.figure(figsize=(100,100))
f, axarr = plt.subplots(1,4)
f.set_figheight(10)
f.set_figwidth(10)
#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])
subplotImage(img3,axarr[2])
subplotImage(img4,axarr[3])
plt.tight_layout()
plt.show()
```

#### Blur
Blurs the image. The amount of blurring is specified by a floating-point value included in the argument.

    python main.py image_dir blur_1.0 blur_2.0 blur_4.0

```python
%run /demo/image_augmentor/main.py /demo/aug_images  blur_1.0 blur_2.0 blur_4.0

img2 = Image.open('/demo/aug_images/bird__blur1.0.png')
img3 = Image.open('/demo/aug_images/bird__blur2.0.png')
img4 = Image.open('/demo/aug_images/bird__blur4.0.png')
# show the images
#plt.figure(figsize=(100,100))
f, axarr = plt.subplots(1,4)
f.set_figheight(10)
f.set_figwidth(10)
#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])
subplotImage(img3,axarr[2])
subplotImage(img4,axarr[3])
plt.tight_layout()
plt.show()
```

#### Compound Example
Produce a single output image for each input image after applying a series for transformations.

    python main.py image_dir  rot_90,fliph

```python
%run /demo/image_augmentor/main.py /demo/aug_images  rot_90,fliph


img2 = Image.open('/demo/aug_images/bird__rot90__fliph.png')
img2 = np.array(img2)

# show the images
f, axarr = plt.subplots(1,2)
f.set_figheight(8)
f.set_figwidth(8)


#orig
subplotImage(img,axarr[0])
# new
subplotImage(img2,axarr[1])

plt.tight_layout()
plt.show()
```
